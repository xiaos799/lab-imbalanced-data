```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("files_for_lab/customer_churn.csv")
df
```

```python
df.info()
```

```python
df['Churn'].replace({'Yes': 1, 'No': 0}, inplace=True)
df['Churn']
```

```python
df.Churn.value_counts()
```

```python
y = df.Churn
X = df[['tenure','SeniorCitizen','MonthlyCharges']]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state=0)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

model = LogisticRegression()
model.fit(X_train, y_train)

pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
print(classification_report(y_test, pred_test))
print(classification_report(y_train, pred_train))
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(k_neighbors=3)
X_train_SMOTE, y_train_SMOTE = sm.fit_resample(X_train, y_train)
X_test_SMOTE, y_test_SMOTE = sm.fit_resample(X_test, y_test)

model = LogisticRegression()
model.fit(X_train_SMOTE, y_train_SMOTE)

pred_train_SMOTE = model.predict(X_train_SMOTE)
pred_test_SMOTE = model.predict(X_test_SMOTE)
print(classification_report(y_test_SMOTE, pred_test_SMOTE))
print(classification_report(y_train_SMOTE, pred_train_SMOTE))
```

```python
from imblearn.under_sampling import TomekLinks

dict= {0: 300, 1: 206}
tl = TomekLinks()

X_train_tl, y_train_tl = tl.fit_resample(X_train, y_train)
X_test_tl, y_test_tl = tl.fit_resample(X_test, y_test)

model = LogisticRegression()
model.fit(X_train_tl, y_train_tl)
model.fit(X_test_tl, y_test_tl)

pred_train_tl = model.predict(X_train_tl)
pred_test_tl = model.predict(X_test_tl)
print(classification_report(y_test_tl, pred_test_tl))
print(classification_report(y_train_tl, pred_train_tl))
```
